## Kiểm tra trùng lặp

```python
duplicate_rows = df[df.duplicated()]
print(f"Số bản ghi trùng hoàn toàn: {duplicate_rows.shape[0]}")
```

```python
df["course_mode_id"].value_counts()
```

```python
duplicate_ids = df[df.duplicated(subset=['id'])]
print(f"Số bản ghi trùng ID: {duplicate_ids.shape[0]}")
```
